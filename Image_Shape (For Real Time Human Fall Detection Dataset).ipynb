```python
import cv2
import os

image_directory = '/content/drive/MyDrive/Fall Detection/train/images'

image_dimensions = None
num_images = 0

for filename in os.listdir(image_directory):
    if filename.endswith('jpg'):
        image_path = os.path.join(image_directory, filename)
        img = cv2.imread(image_path)
        if img is not None:
            num_images += 1
            if image_dimensions is not None and image_dimensions != img.shape:
                print(f"The image {filename} has different dimensions: {img.shape}")
            else:
                image_dimensions = img.shape

print(f"Total Number of images: {num_images}")

if num_images > 0 and image_dimensions is not None:
    print(f"All the images have the same dimensions and same number of channels: {image_dimensions}")
```
